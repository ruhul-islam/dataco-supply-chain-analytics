There are no null values except in Product Description column which is all empty and Order Zipcode which has 155,679 null values. We will delete both columns as they aren't helpful given 100% and 86% of values are missing respectively. We will also delete Customer Password column as it's masked and useless for our analysis.

```python
# Replacing Customer Zipcode missing values with 0 and Customer Lname missing values with Unknown
data_df['Customer Zipcode'] = data_df['Customer Zipcode'].fillna(0)
data_df['Customer Lname'] = data_df['Customer Lname'].fillna('Unknown')

# Verify
data_df.isna().sum()
```

```python
# Deleting 4 useless columns
data_df.drop(columns='Product Description', inplace=True)
data_df.drop(columns='Customer Password', inplace=True)
data_df.drop(columns='Customer Email', inplace=True)
data_df.drop(columns='Order Zipcode', inplace=True)

# Verify
data_df.shape
```

```python
# Confirm no more missing values
data_df.isna().sum()
```

```python
# Check for duplicates
data_df.duplicated().sum()
```

```python
# Change data types of certain columns appropriately
data_df['order date (DateOrders)'] = pd.to_datetime(data_df['order date (DateOrders)'])
data_df['shipping date (DateOrders)'] = pd.to_datetime(data_df['shipping date (DateOrders)'])
data_df['Customer Zipcode'] = data_df['Customer Zipcode'].astype(str)

# Verify
data_df.dtypes
```

```python
# Product Id and Product Category Id are identical
print((data_df['Category Id'] == data_df['Product Category Id']).all())
```

```python
# Delete Product Category Id
data_df = data_df.drop(columns=['Product Category Id'])

# Verify
data_df.shape
```

```python
# Also check Order Item Cardprod Id versus Product Card Id
print((data_df['Order Item Cardprod Id'] == data_df['Product Card Id']).all())
```

```python
# Delete Order Item Cardprod Id
data_df.drop(columns='Order Item Cardprod Id', inplace=True)

# Verify
data_df.shape
```

```python
# Verify with summary
data_df.info()
```

### Feature Engineering

```python
# Delay Days
data_df['delay_days'] = (
    data_df['Days for shipping (real)'] - 
    data_df['Days for shipment (scheduled)']
)

# Verify
data_df['delay_days'].describe()
```

```python
# Binary Flag of if delayed or not where 1 represents delay or 0 represents on time
data_df['is_late'] = (data_df['delay_days'] > 0).astype(int)

 # Verify
data_df['is_late'].unique()
```
